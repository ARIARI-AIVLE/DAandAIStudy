### **Macro F1 score?**
- F1 score
  - precision과 recall의 조화 평균으로 계산되는 모델 성능 평가 지표
  - 주로 이진 분류 문제에서 사용됨
  - 정밀도와 재현율의 균형을 조정하여 정확한 예측과 상승률의 평균을 동시에 고려할 수 있는 장점
<br><br>

- F1 score의 종류
  - **Macro F1 score**
    - 라벨별 f1-score를 산술평균한 것
  - **Weighted F1 score**
    - 라벨별 f1-score를 샘플 수(support)의 비중에 따라 가중평균한 것
  - **Micro F1 score**
    - 전체 TP, FN, FP를 평균한 것이며 이것은 다중 분류에서 정확도와 같다
    - 클래스 불균형이 있는 데이터셋에서 효과적인 평가지표

# Baseline 코드 필사 & 공부

#### Import

- **IsolationForest?**
  - 트리 기반의 비지도 학습 이상치 탐지 모델
  - 랜덤으로 데이터를 split하여 모든 관측치를 고립시키며 구현
    - (정상 데이터보다 이상치를 고립(분리)시키는 것이 더 쉽다!)
  - 특히, 변수가 많은 데이터에서도 효율적으로 작동할 수 있는 장점이 있음
  - 참고: https://velog.io/@vvakki_/Isolation-Forest-%EB%AF%B8%EC%99%84%EC%84%B1

```python
import pandas as pd
import numpy as np

from sklearn.ensemble import IsolationForest
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report

import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings(action='ignore')
```

#### Data Load

```python
data_url = '/content/drive/MyDrive/데이콘_연습/[2] 월간 데이콘 신용카드 사기 거래 탐지 AI 경진대회/'
```

```python
train_df = pd.read_csv(data_url + 'train.csv')
train_df.head()
```

```python
val_df = pd.read_csv(data_url + 'val.csv')
val_df.head()
```

#### Train/Validation Feature 분포 확인

```python
train_df.drop(columns='ID').hist(bins=50, figsize=(20, 20))
plt.show()
```

```python
val_df.drop(columns=['ID']).hist(bins=50, figsize=(20, 20))
plt.show()
```

#### Validation set 사기 거래 비율
- Validation set의 사기 거래 비율이 다른 데이터집합에서도 비슷하게 발생할 것이라고 가정

```python
val_normal, val_fraud = val_df['Class'].value_counts()
val_contamination = val_fraud / val_normal
print(f'Validation contamination : [{val_contamination}]')
```

> 사기 거래는 아주아주 적은 비율을 갖고 있음! (0.1% 정도)

#### Model Define & Fit

```python
# Train dataset은 Label이 존재하지 않음
train_x = train_df.drop(columns=['ID'])    # Input Data
```

```python
# 가설 설정 : Train dataset도 Validation dataset과 동일한 비율로 사기거래가 발생했을 것임.
# -> model parameter : contamination=val_contamination(=0.001055) 적용
model = IsolationForest(n_estimators=125, max_samples=len(train_x), contamination=val_contamination, random_state=42, verbose=0)
model.fit(train_x)
```

#### Evaluation : Validation set

```python
# IsolationForest 모델 출력 (1: 정상, -1: 불량(사기)) 이므로, (0: 정상, 1: 불량(사기))로 Label 변환 시켜주는 함수
def get_pred_label(model_pred):
  model_pred = np.where(model_pred==1, 0, model_pred)
  model_pred = np.where(model_pred==-1, 1, model_pred)

  return model_pred
```

```python
val_x = val_df.drop(columns=['ID', 'Class'])    # Input Data
val_y = val_df['Class']    # Label

val_pred = model.predict(val_x)    # model prediction
val_pred = get_pred_label(val_pred)
val_score = f1_score(val_y, val_pred, average='macro')
print(f'Validation F1 Score :; [{val_score}]')
print(classification_report(val_y, val_pred))
```

#### Inference : Test set

```python
test_df = pd.read_csv(data_url + 'test.csv')    # Test
test_df.head()
```

```python
test_x = test_df.drop(columns=['ID'])
```

```python
test_pred = model.predict(test_x)    # model prediction
test_pred = get_pred_label(test_pred)
```

#### Submission

```python
submit = pd.read_csv(data_url + 'sample_submission.csv')
submit.head()
```

```python
submit['Class'] = test_pred
submit.to_csv(data_url + 'submit.csv', index=False)
```
